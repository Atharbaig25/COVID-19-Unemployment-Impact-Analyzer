```python
df['Month'] = df.iloc[:, 1].str.split('-').str[1].astype(int)

print("Month extraction successful!")
print("Months:", sorted(df['Month'].unique()))
print("\nSample data:")
df[['Region', df.columns[1], 'Month', df.columns[3]]].head()
```

```python
plt.figure(figsize=(12, 6))
monthly_avg = df.groupby('Month')[df.columns[3]].mean()
monthly_avg.plot(marker='o', linewidth=2, color='blue', markersize=8)
plt.title('Unemployment Rate by Month (2020)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.axvspan(4, 6, alpha=0.3, color='red', label='Peak Lockdown')
plt.grid(True, alpha=0.3)
plt.legend()
plt.xticks(range(1, 11))
plt.show()
```

```python
pre_covid = df[df['Month'].isin([1, 2, 3])][df.columns[3]].mean()
during_covid = df[df['Month'].isin([4, 5, 6])][df.columns[3]].mean()
post_covid = df[df['Month'].isin([7, 8, 9, 10])][df.columns[3]].mean()

print("COVID IMPACT:")
print(f"Jan-Mar: {pre_covid:.1f}%")
print(f"Apr-Jun: {during_covid:.1f}%") 
print(f"Jul-Oct: {post_covid:.1f}%")
print(f"Increase: {((during_covid-pre_covid)/pre_covid)*100:.1f}%")
```

```python
plt.figure(figsize=(12, 8))
regional_avg = df.groupby('Region')[df.columns[3]].mean().sort_values(ascending=False)
regional_avg.plot(kind='barh', color='green')
plt.title('Average Unemployment by Region', fontsize=16)
plt.xlabel('Unemployment Rate (%)', fontsize=12)
plt.grid(axis='x', alpha=0.3)
plt.show()

print("TOP 5 WORST REGIONS:")
for i, (region, rate) in enumerate(regional_avg.head().items(), 1):
    print(f"{i}. {region}: {rate:.1f}%")
```

```python
plt.figure(figsize=(10, 6))
region_group = df.groupby(df.columns[6])[df.columns[3]].mean().sort_values(ascending=False)
region_group.plot(kind='bar', color='orange')
plt.title('Unemployment by Geographical Region', fontsize=16)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(data=df, x=df.columns[6], y=df.columns[3])
plt.title('Unemployment Distribution by Region', fontsize=16)
plt.xticks(rotation=45)
plt.grid(True, alpha=0.3)
plt.show()
```

```python
peak_covid = df[df['Month'].isin([4, 5, 6])]
peak_avg = peak_covid.groupby('Region')[df.columns[3]].mean().sort_values(ascending=False)

plt.figure(figsize=(12, 6))
peak_avg.head(8).plot(kind='bar', color='g')
plt.title('Worst Regions During Peak COVID', fontsize=16)
plt.ylabel('Unemployment Rate (%)', fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.show()
```

```python
print("="*50)
print("FINAL SUMMARY")
print("="*50)
print(f"Average: {df[df.columns[3]].mean():.1f}%")
print(f"Max: {df[df.columns[3]].max():.1f}%")
print(f"Min: {df[df.columns[3]].min():.1f}%")

worst = df.groupby('Region')[df.columns[3]].mean().idxmax()
best = df.groupby('Region')[df.columns[3]].mean().idxmin()

print(f"\nWorst: {worst}")
print(f"Best: {best}")
print(f"\nCOVID Peak Impact: {((during_covid-pre_covid)/pre_covid)*100:.1f}% increase")
```
